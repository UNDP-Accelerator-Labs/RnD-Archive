```python
source = 'path/to/{source word doc}'
seed_keywords = [(path.splitext(f)[0], processMD(path.join(source, f)).value) for f in listdir(source) if f != '.DS_Store']
```

```python
def getTypes (md, w):
    principle = 'Purpose/ reason: Why is this important?'
    tactic = r'Tactic description: When do you apply this tactic\? How does it work\?'
    skill = r'Skills description: What does the ability help you do, achieve, or produce\?'
    tool = r'(Type of tool: If it is a tool, what type of tool is it\?|Short description: Describe in one sentence what the tool\/ method is, and what it does.|Description of the tool: what is it\?|Purpose: Why do we use this method or tool for R\&amp;D\? \(max 1,500 characters\))'

    res_principles = len(re.findall(principle, md))
    res_tactics = len(re.findall(tactic, md))
    res_skills = len(re.findall(skill, md))
    res_tools = len(re.findall(tool, md))

    print(w)
    print('pinciples', res_principles)
    print('tactics', res_tactics)
    print('skills', res_skills)
    print('tools', res_tools)
    print('\n')
    
    l = [('principles', res_principles), ('tactics', res_tactics), ('skills', res_skills), ('tools', res_tools)]
    return [t for (t,c) in l if c > 0]
```

```python
def filterMD (md):
    # REMOVE ALL THE CHECKLISTS
    # md = re.sub(r'\#\#\#\# [\w\s\:\?\/\(\)\\;&]+\n', '', md)
    md = re.sub(r'\#\#', '#', md)
    md = re.sub(r'\n-[\s_]+(Sense|Explore|Develop|Test|Diffuse|Catalyze)', '', md)
    md = re.sub(r'\n-[\s_]+(Exploration|Solutions mapping|Experimentation)', '', md)
    md = re.sub(r'\n-[\s_]+(The advocate|The analyst|The communicator|The convener|The hustler|The project manager|The qualitative researcher|The visionary)', '', md)
    md = re.sub(r'\n-[\s_]+(Framework, Model|Guide, Manual|Platform, Portal|Playbook, Toolkit|Worksheet, Canvas|Process|Pilots|Proof of concept|Prototyping)', '', md)
    md = re.sub(r'\n-[\s_]+(Yes|No)', '', md)
    md = re.sub(r'Image: What does it look like\?', '', md)
    md = re.sub(r'Associated trait\(s\)', '', md)
    md = re.sub(r'R\&amp;D Activity: For which R\&amp;D activity is this tool or method used\?', '', md)
    md = re.sub(r'Roles: Which role is this tool or methods associated with\?', '', md)
    md = re.sub(r'_+', ' ', md)
    md = re.sub(r'\\+', '', md)
    # md = re.sub(r'\s+', ' ', md)
    return md.strip()
```

```python
md_content = [(w, filterMD(b), getTypes(b, w)) for (w, b) in seed_keywords]
```

```python
for w,md,t in md_content:
    types = "[[type:{}]]".format(']]\n[[type:'.join(t))
    content = "{}\n\n{}".format(types, md)
    output_name = "{}.md".format(w)
    # TO DO: replace {elements or stories} below to one or the other, depending on the type of file
    output_file = open("../{elements or stories}/pages/{}".format(output_name), 'w')
    output_file.write(content)
    output_file.close()
```
