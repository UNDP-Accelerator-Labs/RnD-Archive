```python
import pandas as pd
```

```python
select_elements = pd.read_excel('data/select_elements.xlsx')
```

```python
id_cols = list(map(lambda x: "id-{}".format(x), list(range(1, 32))))
```

```python
ids = select_elements[~select_elements[id_cols].isnull().all(axis=1)]
```

```python
rm_duplicates = ids.duplicated(subset=id_cols, keep='last')
unique_ids = ids[~rm_duplicates].sort_values(by=id_cols)
unique_ids.reset_index(inplace=True)
```

## Get the pool of pad ids to pull from for the classification of stories

```python
pool_ids = unique_ids[id_cols]
pool_ids = pool_ids.values[~pd.isna(pool_ids.values)].tolist()
pool_ids = [int(n) for n in pool_ids]
```
